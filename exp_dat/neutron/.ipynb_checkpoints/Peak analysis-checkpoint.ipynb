# Test condition

```python
# Stress data file
stress=np.loadtxt('Strain_BCC.dat',skiprows=1).T[2]

# Wavelength of neutron
wl=1.47*1e-9

# the data file folders
directories = ['ss111_fe110','fe200','fe112','fe220','ss111_fe110','ss200','ss311','ss222']
hkl        = [r'$(110)_\alpha$',r'$(200)_\alpha$',r'$(112)_\alpha$',r'$(220)_\alpha$',
               r'$(111)_\gamma$',r'$(200)_\gamma$',r'$(311)_\gamma$',r'$(222)_\gamma$']

# The X_c values (2theta)
Xc=[41,60.4,76.6,91.4,41,47.2,84,88.4]

def conv(Xc):
    """ The channel's id into theta for the given Xc"""
    ch=np.linspace(1,1024,1024)
    #ch=np.array(ch,dtype='int')
    ch = Xc+0.35+(400-ch)*0.00826
    return ch
```

# Necessary modules

- to remove the trivial intensity data

```python
def trim(x):
    """ to trim the trivial data points in '.txt' files """
    i0=0; i1=-1    
    for i in range(len(x)):
        if x[i]!=0: 
            i0=i
            break
    for i in range(len(x)):
        if x[::-1][i]!=0:
            i1=i
            break
    i1=len(x)-i
    return i0,i1
```

- Convert th2 to d-spacing to strain based on the Bragg's law

$n\lambda = 2d\sin\theta $

$d = \frac{n\lambda}{2\sin{\theta}}$

$d^1 = \frac{n\lambda}{2\sin({\theta+\theta^{\mathrm{e}}})}, d^2 = \frac{n\lambda}{2\sin({\theta-\theta^{\mathrm{e}}})}$

- Meanwhile, the lattice strain id defined:

$\varepsilon = d/d_0 - 1$

$\varepsilon^1 = (d^1)/d_0 - 1,  \varepsilon^2 = (d^2)/d_0 - 1$

$2\Delta \varepsilon = \varepsilon^2 - \varepsilon^1  = (d^2-d^1)/d_0 = \frac{n\lambda}{2} [1/\sin(\theta+\theta^e) - 1/\sin(\theta-\theta^e) ]$

```python
def th2_to_eps(th2,var,wl):
    """ convert 2theta to d-spacing; 

    nlamb = 2d sin(theta) 
    lamb/(2d) = sin(theta) 
    d = lamb/ sin(theta)
    
    Arguments
    th2 = 2 theta
    var = error
    wl  = wavelength
    """
    dspacings = []
    dspacings_error=[]
    th2rad = np.pi/180.
    for i in range(len(th2)):
        
        # calc d-spacings
        d  = 0.5 * wl / np.sin(th2[i]/2. * th2rad)
        d1 = 0.5 * wl / np.sin((th2[i]+var[i])/2. * th2rad) #d+
        d2 = 0.5 * wl / np.sin((th2[i]-var[i])/2. * th2rad) #d-
        dd = abs(d1-d2) # |(d+) - (d-)|
        
        dspacings.append(d)
        dspacings_error.append(dd)
        
    dspacings       = np.array(dspacings)
    dspacings_error = np.array(dspacings_error)
    
    eps = dspacings/dspacings[0] - 1.
    deps = dspacings_error/dspacings[0]/2.
    
    return eps,deps
        
```

- PP reads and returns peak profiles for a given 'id' file

```python
def pp(id=0,nstp=1,off=0,iplot=False):
    name=directories[id]
    fk=glob('0312_MV8/detail/%s/*.kis'%directories[id])
    ft=glob('0312_MV8/detail/%s/*.txt'%directories[id])
    ft.sort(); fk.sort()    

    def_exptime = 60000.
    
    th2=conv(Xc[id]) # Generate suitable 2theta axis

    ys = []; xs = []
    for i in range(len(fk)):
        nf=i
        if i%nstp==0:
            fn = fk[nf].split('.kis')[0]
            f = os.popen('head -21 %s.kis'%(fn),'r')
            lines=f.readlines()
            f.close()

            # start th0; th0=float(lines[16].split('=')[1].split('[')[0])            

            # exposure time
            expt = float(lines[9].split('=')[1].split('(')[0])
            fn = ft[nf].split('.txt')[0]
            intensity = np.loadtxt('%s.txt'%fn)
            
            # trim
            i0,i1=trim(intensity)
            y = intensity*def_exptime/expt # normalize the intensity based on the exposure time
            y = y[i0:i1];  x = th2[i0:i1]

            # It is in descending order
            x = x[::-1]; y = y[::-1]
            ys.append(y); xs.append(x)
            if iplot: plot(x,y+i*off,'+',label='%s'%(i+1))

    if iplot:
        #legend(loc='best',fontsize=9)
        #title(name)
        grid('on')
        gca().set_xlabel(r'2$\theta$',dict(fontsize=20))
        gca().set_ylabel('Intensity',dict(fontsize=18))
    
    return xs, ys


def pp2f(id=0,k=0):
    """ write the pp'ed data to files """
    fn = 'profile_id%i_%i.txt'%(id,k)
    
```

- Curve fit to Gaussian distribution function

```python
from scipy.optimize import curve_fit
# Gaussian definition
def gauss(x, *p):
    A, mu, sigma = p
    return A*numpy.exp(-(x-mu)**2/(2.*sigma**2))
def fit(x,y):
    """
    1. Fit a peak (x,y) to a Gaussian function
    2. Returns popt and pcov from scopy.optimize.curve_fit
        # popt: Optimal values for the Gaussian function parameters (A,mean,sigma)
    """
    x=np.array(x); y=np.array(y)
    n=len(x)

    # the peak location is estimated
    ix=np.where(y==max(y))
    # the center of the peak
    mean = x[ix][0]
    # standard deviation
    sigma = sum(y*(x-mean)**2)/n 
    ix=np.where(y==max(y))

    # initial guess for curve_fit
    p0 = [max(y),x[ix],sigma**2]
    #p0 = [max(y),x[ix],1]
    popt,pconv = curve_fit(gauss,x,y,p0)
    return popt, pconv
```

- Given the id folder and k, read the peaks and fit them with Gaussian


-- Write the 2theta-counts profiles data to files

-- Return the fitted mean value (2$\theta$) and its variants

```python
def th2_gauss(id=5,k=1,ifig=None,iexp=True,c='r'):
    """ Read the peak profiles belonging to the give id folder (hkl) of k-th scan
    """
    if ifig!=None: plt.figure(ifig)
    xs,ys=pp(id=id,nstp=1,off=0)

    # a trick is that I reduce the amount of raw data points 
    # to work-around in case that the curve_fit is not successful.
    nstp = 1
    ikeep=True
    while(ikeep):
        y=ys[k][::nstp]; x=xs[k][::nstp]
        try:
            popt, pconv = fit(x,y)
        except:
            nstp = nstp + 1 # trick to work-around the fit failure
        else:
            #print 'mean', popt[1]
            #print 'std', popt[2]
            if ifig!=None and iexp: plot(x,y,'kx',alpha=0.5)
            yfit = gauss(x,*popt)
            
            # integrate the peak ... to find the intensity
            dx = np.diff(x)
            dx=dx>0
            if not(all(dx==True)):
                raise IOError,'dx is not monotonically increasing...'
            
            I=np.trapz(x[::],yfit[::])
            
            if ifig!=None: plot(x,yfit,'%s-'%c)
            if ifig!=None: gca().set_ylim(-1,); grid('on')
            ikeep=False

    return popt[1],popt[2],I,x
```

```python
## Read hkl (id=4) of 8-th scan
mean,std,I,x=th2_gauss(id=4,k=8,ifig=True)
gca().set_title(hkl[4],dict(fontsize=20))
print 'theta_b', mean, std
```

```python
## Read hkl (id=4) of 8-th scan
def total_prof(k=0):
    m=[True,True,True,True,False,True,True,True]
    c=['r','r','r','r','b','b','b','b']    
    for i in range(8):
        if m[i]==True:
            mean,std,I,x=th2_gauss(id=i,k=k,ifig=1,iexp=False,c=c[i]);

total_prof(k=0)
total_prof(k=1)
total_prof(k=2)
total_prof(k=3)


#gca().set_title(hkl[4],dict(fontsize=20))
#gca().set_ylim(0,50)
gca().set_xlabel(r'$2\theta$',dict(fontsize=18))
gca().set_ylabel('Counts',dict(fontsize=18))
print 'theta_b', mean, std
```

```python
def dspacings(id=3,ifig0=10):
    n=np.arange(0,9)
    
    th2s=[]; var=[]; Ints=[]
    for i in range(len(n)):
        if ifig0!=None: ifig=ifig0+i
        elif ifig0==None: ifig= None
        mean,std,I,x = th2_gauss(id,i,ifig=ifig)
        th2s.append(mean)
        var.append(std)
        Ints.append(I)
        #print mean
        
    return np.array(th2s),np.array(var),np.array(Ints)

th2s,var,Ints=dspacings(id=5,ifig0=1)
```

# Example

```python
def main(id=5):    
    th2s,var,Ints=dspacings(id=id,ifig0=None)
    eps,epse=th2_to_eps(th2s,var,wl)
    
    fig1=plt.figure(1,figsize=(6,10))
    ax1=fig1.add_subplot(211)
    ax2=fig1.add_subplot(212)
    ax1.plot(stress,eps[:-1]*1e6,'-x',label=hkl[id])
    #errorbar(stress,eps[:-1]*1e6,yerr=epse[:-1]*1e6,fmt='-x',label=hkl[id])
    ax1.legend(loc='best',fancybox=True).get_frame().set_alpha(0.5)
    ax1.grid('on')
    ax1.set_ylabel(r'$[\mu\varepsilon]$', dict(fontsize=18))
    ax1.set_xlabel(r'$\sigma$ [MPa]', dict(fontsize=18))
    
    
    ax2.plot(stress,Ints[:-1],'-x',label=hkl[id])
    #ax2.legend(loc='best',fancybox=True).get_frame().set_alpha(0.5)
    ax2.grid('on')
    ax2.set_ylabel('Intensity', dict(fontsize=18))
    ax2.set_xlabel(r'$\sigma$ [MPa]', dict(fontsize=18))
    
    return stress,eps[:-1],Ints[:-1]
```

```python
stress,eps,ints=main(id=3)
```
