* Plotting elastic strain from neutron diffraction

```python
import os
from glob import glob
fns = glob('*.dat')
print fns
```

* BCC

```python
f=os.popen('head S*BCC.dat','r')
lines=f.readlines()
print lines[0]
print lines[-1]
```

```python
dat=np.loadtxt(fns[0],skiprows=1)
dat=dat.T
print dat.shape
x=dat[2]
errorbar(x,dat[3],yerr=dat[4],fmt='x--',label='(220)')
errorbar(x,dat[5],yerr=dat[6],fmt='o--',label='(211)')
errorbar(x,dat[7],yerr=dat[8],fmt='+--',label='(200)')
plot(x,dat[9],'.--',label='(110)')
#errorbar(x,dat[9],yerr=dat[10],fmt='*-',label='(110)')
legend(loc='best',fancybox=True).get_frame().set_alpha(0.5)
gca().set_xlabel('Stress [MPa]',dict(fontsize=20))
gca().set_ylabel(r'Elastic strain [$\mu$]',dict(fontsize=20))
grid('on')
```

* for FCC

```python
dat=np.loadtxt(fns[1],skiprows=1,dtype='str')
dat = np.genfromtxt(fns[1],skiprows=1)
dat=dat.T
x=dat[2]
errorbar(x,dat[3],yerr=dat[4],fmt='x--',label='(111)')
errorbar(x,dat[5],yerr=dat[6],fmt='o--',label='(200)')
errorbar(x,dat[7],yerr=dat[8],fmt='+--',label='(311)')
plot(x,dat[9],'.--',label='(222)')
#errorbar(x,dat[9],yerr=dat[10],fmt='*-',label='(110)')
legend(loc='best',fancybox=True).get_frame().set_alpha(0.5)
gca().set_xlabel('Stress [MPa]',dict(fontsize=20))
gca().set_ylabel(r'Elastic strain [$\mu$]',dict(fontsize=20))
grid('on')
```

* Together

```python
dat=np.loadtxt(fns[0],skiprows=1)
dat=dat.T

print dat.shape
x=dat[2]

errorbar(x,dat[3],yerr=dat[4],fmt='x-',label=r'(220)$_\alpha$')
errorbar(x,dat[5],yerr=dat[6],fmt='o-',label=r'(211)$_\alpha$')
errorbar(x,dat[7],yerr=dat[8],fmt='+-',label=r'(200)$_\alpha$')
plot(x,dat[9],'.-',label=r'(110)$_\alpha$')


dat=np.loadtxt(fns[1],skiprows=1,dtype='str')
dat = np.genfromtxt(fns[1],skiprows=1)
dat=dat.T
x=dat[2]
errorbar(x,dat[3],yerr=dat[4],fmt='x--',label=r'(111)$_\gamma$')
errorbar(x,dat[5],yerr=dat[6],fmt='o--',label=r'(200)$_\gamma$')
errorbar(x,dat[7],yerr=dat[8],fmt='+--',label=r'(311)$_\gamma$')
plot(x,dat[9],'.--',label=r'(222)$_\gamma$')
#errorbar(x,dat[9],yerr=dat[10],fmt='*-',label='(110)')
legend(loc='best',fancybox=True).get_frame().set_alpha(0.5)
gca().set_xlabel('Stress [MPa]',dict(fontsize=20))
gca().set_ylabel(r'Elastic strain [$\mu$]',dict(fontsize=20))
gca().grid('on')
gca().set_ylim(0,12000)
```
